## Question 3 solution
For every country, find the top-5 weeks with the highest number of excess deaths.

```python
# Create a new DataFrame that includes the following columns:
# location, death_excess, week
df1 = data[['location','Week','deaths_2020_all_ages','deaths_2019_all_ages']]
```

```python
df1 = df1.dropna()
```

```python
df1['death_excess'] = df1['deaths_2020_all_ages'] - df1['deaths_2019_all_ages']
```

```python
# Create a subset
df_subset = df1[['location','Week','death_excess']]
```

```python
df_subset.head(5)
```

```python
# Group countries and sort death_excess in a descending order and extract the top 5 values for each country
# replaces the previous DataFrame index with the new integer index provided by reset_index() 
df_max = df_subset.groupby('location').apply(lambda x: x.sort_values(by='death_excess',ascending=False).head(5)).reset_index(drop=True)
```

```python
df_max
```

```python
# Take a look at the df_max
# Show all rows, the function suggested by the link
# https://dev.to/chanduthedev/how-to-display-all-rows-from-data-frame-using-pandas-dha
pd.set_option('display.max_rows', None)  
df_max
```

From the result above, we can find the top-5 weeks with the highest number of excess deaths for each country. For example, Austria reached the top 5 highest excess deaths in Week 49, 47, 48, 50, 46 with death_excess numbers 948, 859, 855, 807 and 727 respectively. The week information is an indication of month and season in a year as well, normally Winter and early Spring are the outbreak period of flu. The example of Austria peak weeks are around the end of the year (before and after Christmas time, 50 * 7 = 350). 

## Question 4 solution
For each country, find which year had the lowest mortality.

```python
# Get columns from the original dataset == > data 
df2 = data[['location','deaths_2010_all_ages','deaths_2011_all_ages','deaths_2012_all_ages','deaths_2013_all_ages',
           'deaths_2014_all_ages','deaths_2015_all_ages','deaths_2016_all_ages','deaths_2017_all_ages',
           'deaths_2018_all_ages','deaths_2019_all_ages','deaths_2020_all_ages','deaths_2021_all_ages']]
```

```python
# Fill missing values with 0 
df2 = df2.fillna(0)
```

```python
# Sum up the mortality of each year of each country
df2 = df2.groupby('location').sum()
```

```python
#reset the index (by default using ordered number index)
df2.reset_index(inplace=True)
```

```python
# Take a look at the first 5 rows of df2
df2.head(5)
```

```python
# Extract the location column and put the data in a list
location = []
for i in df2['location']:
    location.append(i)
```

```python
# Drop the column location (prepare for getting the lowest value of each row later)
df_data = df2.drop(columns=['location'])
```

```python
# calculate the minimum value along each row of the DataFrame ==> df_data
# Function suggested by the link
# https://www.geeksforgeeks.org/how-to-get-the-minimum-value-from-the-pandas-dataframe-in-python/
min_rows = df_data.min(axis=1)
```

```python
# Get the corresponding column name of the minimum value of each row
# Function suggested by the link
# https://www.geeksforgeeks.org/python-pandas-dataframe-idxmin/?ref=header_search
min_columns = df_data.idxmin(axis=1)
```

```python
type(min_rows) # series data type
```

```python
type(min_columns) # series data type
```

```python
# Write a DataFrame includes 3 columns (location, Year, lowest_mortality)
result_df = pd.DataFrame({'location':location,'year': min_columns,'lowest_mortality': min_rows})
```

```python
# Take a look at the result
result_df
```

Cross check with the result from df2 and we can see for example the lowest value for Austria was from 2021 with a number of 3885, the same number of 3885 from result_df. The result_df also shows some places had the lowest 0 (this might be because of missing value). The lowest values occurred mostly in 2021 for each location and this may be due to the end of the COVID-19 period.
